```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('/content/green_tree.jpg',cv2.IMREAD_UNCHANGED)
```

```python
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(cv2.cvtColor(gray, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
ret, thresh = cv2.threshold(gray, 150, 255, cv2.THRESH_BINARY_INV)
```

```python
contours, hierarchy = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
```

```python
for c in contours:

    x,y,w,h = cv2.boundingRect(c)
    cv2.rectangle(image, (x,y), (x+w, y+h), (255, 0, 0), 10)

    rect = cv2.minAreaRect(c)

    box = cv2.boxPoints(rect)
    #box = int(box)  # Replace np.int0 with int
    box = box.astype(int) # Convert the array elements to integers

    cv2.drawContours(image, [box], 0, (0,0, 255), 10)
```
